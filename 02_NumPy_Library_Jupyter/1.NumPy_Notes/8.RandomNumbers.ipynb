```python
# Random Numbers with Integer Values

import numpy as np

# Create a NumPy array or random integers between 0 and 10

random_intergers = np.random.randint(0,11,size = 10)
print(f"Random Integers: \n {random_intergers}")

# Create a NumPy array of random integers between 0 and 10 with shape (3,3)

rand_int = np.random.randint(0,11,size = (3,3))
print(f"Random Integers with shape (3,3):\n {rand_int}")
```

```python
# Random Numbers with Float Values

import numpy as np

# Create a NumPy array of random float values between 0 and 1

rand_float = np.random.rand(5)
print(f"Random Float Values: \n {rand_float}")

# Create a NumPy array of random float values with shape (2,4)
rand_float1 = np.random.uniform(0,1,size = (2,4))
print(f"Random Float Values with shape (2,4):\n {rand_float1}")
```

```python
import numpy as np

# Create a NumPy array of random integers between 1 and 100

rand1 = np.random.rand(10) 
# here rand will generate the number of values according to the given size between 0 and 1.
print(f"The Value of random float with rand is : \n{rand1}")

rand2 = np.random.uniform(1,10,10)
# here uniform will generate the number of values according to the size in between the given range. 
print(f"The Value of random float with uniform is : \n{rand2}")
```

```python
# To Shuffle the random generated number.

import numpy as np

array = np.random.randint(1,11,5)
print(f"The array of random integers is: \n {array}")

# Shuffle

np.random.shuffle(array)
print(f"The shuffled array is : \n {array}")
```

```python
# To print the choice of the fruit

import numpy as np

fruits = np.array(['apple','banana','mango','cherry','guavava'])

print(f"The fruit are :\n{fruits}")

fruit = np.random.choice(fruits)
print(f"The fruit of own choice is: \n{fruit}")
```
