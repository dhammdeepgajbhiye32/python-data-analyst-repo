```python
# Introduction to NumPy Multidimensional Arrays Like = 0D,1D,2D,3D Arrays
```

```python
import numpy as np  
array = np.array(['A'])                             # Creating a 0D array

print("Multidimensional Array:\n", array)           # Displaying the array
print("Array Dimension:", array.ndim)               # Displaying the number of dimensions
print("Array Shape:", array.shape)                  # Displaying the shape of the array
```

```python
import numpy as np

array = np.array(['A','B','C'])                             # Creating a 1D array

print("Array:\n", array)                            # Displaying the array
print("Array Dimension:", array.ndim)               # Displaying the number of dimensions
print("Array Shape:", array.shape)                  # Displaying the shape of the array
```

```python
import numpy as np

array = np.array([['A', 'B','C'],                   # Creating a 2D array
                  ['D', 'E','F'],
                  ['G', 'H','I']])
print("Array:\n", array)                            # Displaying the array
print("Array Dimension:", array.ndim)               # Displaying the number of dimensions
print("Array Shape:", array.shape)                  # Displaying the shape of the array
```

```python
import numpy as np

array = np.array([[['A','B','C'],['D','E','F'],['G','H','I']],      # Creating a 3D array
                  [['J','K','L'],['M','N','O'],['P','Q','R']],
                  [['S','T','U'],['V','W','X'],['Y','Z','AA']]])        

print("Array:\n", array)                            # Displaying the array  
print("Array Dimension:", array.ndim)               # Displaying the number of dimensions
print("Array Shape:", array.shape)                  # Displaying the shape of the array
```
