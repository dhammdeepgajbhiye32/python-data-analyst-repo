```python
# Here we will see how to use AGGREGATE FUNCTIONS in Python with NumPy.
```

```python
import numpy as np

array = np.array([[1, 2, 3], [4, 5, 6]])

# Sum of all elements
total_sum = np.sum(array)
print("Total Sum:", total_sum)
print("\n")

# Mean of all elements
mean_value = np.mean(array)
print("Mean Value:", mean_value)
print("\n")

# Standard Deviation of all elements
stc_dev = np.std(array)
print("Standard Deviation:", stc_dev)
print("\n")

# Minimum value in the array
min_value = np.min(array)
print("Minimum Value:", min_value)
print("\n")

# Maximum value in the array
max_value = np.max(array)
print("Maximum Value:", max_value)
print("\n")

# Sum along columns (axis=0)
sum_columns = np.sum(array, axis=0)
print("Sum along columns:", sum_columns)
print("\n")

# Sum along rows (axis=1)
sum_rows = np.sum(array, axis=1)
print("Sum along rows:", sum_rows)
```
