```python
# This project is Student Marks Analysis.

# Objective
# 1. Generate marks of 100 students(0-100)
# 2. Find mean, median, top scorers, failure count(<35)
# 3. Normaliz marks for curve grading.
```

```python
import numpy as np

# Step 1: Generate marks of 100 students(0-100)

marks = np.random.randint(0,101, size = 100)
print(f"Marks of 100 students : \n {marks}")

# 1. Finding mean

mean_marks = np.mean(marks)
print(f"Mean Marks : {mean_marks}")

# 2. Finding median

median_marks = np.median(marks)
print(f"Median Marks : {median_marks}")

# 3. Finding top scorers

max_marks = np.max(marks)
print(f"Top scorers Marks : {max_marks}")

# 4. Finding lowest scorers
min_marks = np.min(marks)
print(f"Lowest scorers Marks : {min_marks}")

# 5. Finding failure count(<35)
fail_count = np.sum(marks < 35)
print(f"Failure Count (<35) : {fail_count}")

# 6. Normalizing marks for curve grading.
normalized_marks = (marks - np.min(marks)) / (np.max(marks) - np.min(marks)) * 100
print(f"Normalized Marks : \n {normalized_marks}")

# End of the project
```
