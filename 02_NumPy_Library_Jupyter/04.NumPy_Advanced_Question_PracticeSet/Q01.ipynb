```python
# Quesiton 1 : Create a Random Dataset of 100 samples and 
# split them into Train with (80%) and Test with (20%).
```

```python
import numpy as np
array = np.random.randint(0,100,size = 100)
print(f"Random Dataset of 100 samples:\n{array}\n")

train = array[:80]                          # 0 - 80
print(f"Train Dataset (80%) :\n{train}\n")

test = array[80:]                           # 80 - 100
print(f"Test Dataset (20%) :\n{test}")
```
