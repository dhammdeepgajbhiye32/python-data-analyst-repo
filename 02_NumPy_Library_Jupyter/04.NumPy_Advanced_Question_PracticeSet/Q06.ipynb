```python
# Question : # Quesiton: Student Marks Analysis System using NumPy 
# Project Description:
# Create a Student Marks Analysis System using NumPy that allows users to input student names and
# their marks in various subjects. The system should be able to perform the following operations:
# 1. Add Student Marks: Input student names and their corresponding marks in subjects 
#    like Maths, Science, and English.
# 2. Calculate Average Marks: Compute the average marks for each student across all subjects.
# 3. Subject-wise Analysis: Calculate the subject wise highest score across all students.
# 4. Find Top Scorer: Identify the student with the highest average marks.
# 5. Implement Pass/Fail logic using conditonal statements.
```

```python
import numpy as np

# Create an array to hold student names and their marks
# 1. Adding names of students
student_names = ["Alice", "Bob", "Charlie", "David"]

print("Student Names:")
for i in range(len(student_names)):
    print(f"{student_names[i]}")

# Marks in subjects: Maths, Science, English
marks = np.array([[85, 90, 78],   # Alice
                  [88, 76, 92],   # Bob
                  [90, 91, 85],   # Charlie
                  [70, 65, 80]])  # David

print("\n1. Student Marks (Maths, Science, English):")
for i in range(len(student_names)):
    print(f"{student_names[i]}: {marks[i]}")

# 2. Calculate Average Marks for each student

avg_marks = np.mean(marks,axis = 1)
print("\n2. Average Marks for each student:")
for i in range (len(student_names)):
    print(f"{student_names[i]}: {avg_marks[i]:.2f}")

# 3. Subject-wise Analysis: Calculate the subject wise highest score across all students.

highest_score = np.max(marks,axis = 0)
subject = ["Maths", "Science", "English"]
print("\n3. Subject-wise Highest Scores:")
for i in range(len(subject)):
    print(f" {student_names[i]} : {subject[i]} : {highest_score[i]}")


# 4. Find Top Scorer: Identify the student with the highest average marks.

top_scorer_index = np.argmax(avg_marks)   # index of the student with highest average marks
print(f"\n4. Top Scorer: {student_names[top_scorer_index]} with average marks of {avg_marks[top_scorer_index]:.2f}")


# 5. Implement Pass/Fail logic using conditional statements.
pass_mark = 60
print("\n5. Pass/Fail Status:")
for i in range(len(student_names)):
    if np.all(marks[i] >= pass_mark):
        status = "Pass"
    else:
        status = "Fail"
    print(f"{student_names[i]}: {status}")
```
