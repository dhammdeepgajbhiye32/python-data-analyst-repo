```python
# Question 5 : Create a synthetic dataset 1000 rows and 
# analyze the missing percentage.
```

```python
import numpy as np

array = np.random.rand(100)
array[np.random.randint(0,100,size = 50)] = np.nan
print(f"Array with NaNs: {np.round(array,2)}")

nan_count = np.isnan(array).sum()
print(f"Number of NaNs: {nan_count}")

nan_per = (nan_count / array.size ) * 100
print(f"Percentage of NaNs: {nan_per}%")
```
