```python
# Question 4 : Perform the Broadcasting operation on two numpy arrays of different shapes.
```

```python
import numpy as np

# Creating a 2 sample arrays of different shapes
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])

# Performing broadcasting operation
c = a * b
print(f"Array c after broadcasting:\n{c}")
```
