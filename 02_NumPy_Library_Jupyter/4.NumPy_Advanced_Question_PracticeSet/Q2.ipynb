```python
# Question 2 : Extract the values in between 30 and 60 in the given array.
```

```python
import numpy as np

# Creating a sample array

array = np.random.randint(10,100, size = 100)

print(f"Original Array: {array}")

# # Extracting values between 30 and 60
extract_values = array[((array > 30) & (array < 60))]
print(f"The Extraced values are: \n {extract_values}")
print(f"The length of the extracted values are: {len(extract_values)}")
```
