```python
# Question 3 : Remove all NaN values from the given array.
```

```python
import numpy as np

# Create a simple array

array = np.array([1, 2, np.nan, 4, np.nan, 5])

print(f"Original Array: \n {array}")

# Removing all NaN values from the array

clean_array = array[~np.isnan(array)]
print(f"Array after removing NaN values: \n {clean_array}")
```
