```python
# Question 5 : Perform Normalising a given array using Min-Max Scaling.
```

```python
import numpy as np

# Create a sample array

array = np.array([10,20,30,40,50])
print(f"The original array is: \n {array}")

# Performing Normalizatio using Min-Max Scaling

normalized_array = (array - np.min(array)) / (np.max(array) - np.min(array))

print(f"The Normalized array using Min-Max Scaling is: \n {normalized_array}")
```
