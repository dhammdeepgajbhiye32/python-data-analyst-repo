```python
# Without using NumPy, multiply each element in a list by 2

my_list = [1, 2, 3, 4, 5]
print("Original list:", my_list)

# After multiplying each element by 2
my_list = my_list * 2
print("After multiplying by 2:", my_list)
```

```python
# By using NumPy, multiply each element in an array by 2

import numpy as np

array  = np.array([1, 2, 3, 4, 5])
print("Original array:", array)

# After multiplying each element by 2
array = array * 2
print("After multiplying by 2:", array) 
```
