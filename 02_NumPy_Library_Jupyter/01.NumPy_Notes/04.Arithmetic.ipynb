```python
import numpy as np

# Scalar Arithmetic Operations

array = np.array([10,20,30,40,50])

print(f"Original Array: {array}")

# Addition
add = array + 5
print(f"After Addition: {add}")

# Subtraction
sub = array - 5 
print(f"After Subtraction: {sub}")

# Multiplication
mul = array * 2
print(f"After Multiplication: {mul}")

# Division
div = array / 2
print(f"After Division: {div}")

# Modulus
mod = array % 3
print(f"After Modulus: {mod}")

# Exponentiation
exp = array ** 2
print(f"After Exponentiation: {exp}")

# Floor Division
floordiv = array // 3
print(f"After Floor Division: {floordiv}")
```

```python
import numpy as np

# Vectorized Arithmetic Operations

array = np.array([10,20,30,40,50])

print(f"Original Array: {array}")

print(f"Square Root: {np.sqrt(array)}")  # Square Root

array2 = np.array([1.01,2.26,3.56,4.75,5.95])

print(f"Rounded Array: {np.round(array2)}")  # Rounding Off
```

```python
# Question: Calculate the Area of a Circle with given Radii
radii = np.array([1,2,3,4,5])

print(f"Radii: {radii}")                # Radii of Circles

area = np.pi * (radii ** 2)

print(f"Area of Circles: {area}")       # Area of Circles
```

```python
# Eleements-wise Operations Between Two Arrays

import numpy as np

array1 = np.array([1,2,3])
array2 = np.array([4,5,6])

# addition
add = array1 + array2
print(f"After Addition: {add}")

# subtraction
sub = array1 - array2
print(f"After Subtraction: {sub}")  

# multiplication
mul = array1 * array2
print(f"After Multiplication: {mul}")

# division
div = array1 / array2
print(f"After Division: {div}")

# modulus
mod = array1 % array2
print(f"After Modulus: {mod}")

# exponentiation
exp = array1 ** array2
print(f"After Exponentiation: {exp}")

# floor division
floordiv = array1 // array2
print(f"After Floor Division: {floordiv}")
```

```python
# Comparison Operations Between Two Arrays

import numpy as np

array = np.array([10,20,30,40,50])

print(f"Original Array: {array}")

print(f"Equal to 30: {array == 30}")      # Equal to 30

print(f"Greater than 25: {array > 25}")          # Greater than 25  

print(f"Less than or equal to 40: {array <= 40}")  # Less than or equal to 40

print(f"Not equal to 20: {array != 20}")      # Not equal to 20

print(f"Greater than or equal to 15: {array >= 15}") # Greater than or equal to 15

print(f"Less than 50: {array < 50}")         # Less than 50
```
