```python
# Here we will see how to save and load the NumPy arrays.
```

```python
# To Save and Load a single array.

import numpy as np

data = np.array([1,2,3,4,5])

# To save the NumPy Array we use save.npy method.

np.save("dataFile",data)
print("The Data NumPy array is successfully saved in dataFile.")

# To Load the NumPy Array we use load method

database = np.load("dataFile.npy")
print("The dataFile.npy file is successfully Loaded in database Array")
```

```python
# To save and load the multiple NumPy Arrays

import numpy as np

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])

# To save the multiple array we use save.npz method

np.savez("dataFile1",arr1,arr2)
print("Both the arrays are saved in dataFile1")

# To load the multiple array we use load.npz method

dataBase = np.load("dataFile1.npz")
print("The dataFile1.npz file is successfully Loaded in dataBase Array")
print(dataBase)

array1 = dataBase["arr_0"]
print(f"array1 is : \n {array1}")

array2 = dataBase["arr_1"]
print(f"array2 is : \n {array2}")
```
