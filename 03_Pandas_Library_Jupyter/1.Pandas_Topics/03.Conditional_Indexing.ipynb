```python
# Creating a series using dictionary

import pandas as pd

fruit_proteins = {
    "apple": 0.3,           # protein in grams
    "banana": 1.1,  
    "mango": 0.8,
    "orange": 0.9,
    "grape": 0.7,
    "kiwi": 1.0,
    "pineapple": 0.5,
    "peach": 0.6,
    "cherry": 1.0,
    "watermelon": 0.4
}

data = pd.Series(fruit_proteins,name="Proteins (g)")
print(f"Fruit Proteins Series:\n{data}\n")

print(f"Fruits with protein content greater than 0.5 grams:\n{data[data>0.5]}")

```

```python
# Performing AND, OR, and NOT operations on Series

# AND operation

print(f"\n Fruits with protein content greater than 0.5 grams AND less than 1.0 gram : \n{data[(data > 0.5) & (data < 1.0)]}")

# OR operation

print(f"\n Fruits with protein content less than 1.0 grams OR greater than 0.5 gram : \n {data[(data < 1.0) | (data > 0.5)]}")

# NOT operation

print(f"\n Fruits with protein content NOT greater than 1.0 grams : \n {data[~(data > 1.0)]}")
print(f"\n Fruits with protein content NOT Less than 0.5 grams : \n {data[~(data < 0.5)]}")
```
