```python
# To perform the Broadcasting operation on DataFrame

import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, np.nan, 45],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Sales'],
    'Salary': [50000, 60000, 70000, np.nan, 90000]}

# Converting a dictionary to a DataFrame

df = pd.DataFrame(data)
print("\n Original Dataframe is: \n")
df
```

```python
# To perform the Broadcasting operation on DataFrame

df['Salary']  = df['Salary'] + 5000  # Here we increase each salary by 5000
df
```
