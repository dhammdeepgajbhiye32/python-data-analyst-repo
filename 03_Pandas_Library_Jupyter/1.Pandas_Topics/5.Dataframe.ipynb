```python
import numpy as np
import pandas as pd

data = {
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Age" : [24, 30, np.nan, 35, 28, 40],
    "Department" : ["HR", "Finance", "IT", "IT", "Finance", "HR"],
    "Salary" : [50000, 60000, 55000, 70000, np.nan, 80000]
}

data

# Converitng a dictionary to a DataFrame

df = pd.DataFrame(data)

print(f"\n Original DataFrame:\n{df}")

dataframe = df.copy() # Creating a copy of the original DataFrame

print(f"\n First 3 rows of the DataFrame: \n{df.head(3)}")
print(f"\n Last 2 rows of the DataFrame: \n{df.tail(2)}")
```

```python
# Applying iloc method in DataFrame

print(f"\n Accessing specific rows using iloc : \n {df.iloc[1:3]}")
# It accesses rows at index 1 and 2 (excluding index 3)

# To access specific rows and columns we need to use loc method

# Accessing specific rows and columns using loc
print(f"\n Accessing specific rows and columns using loc : \n {df.loc[1:3,["Name","Age","Salary"]]}")

# Accessing columns using loc method
print(f"\n Accessing all rows with specific columns using loc : \n {df.loc[:,["Name","Department"]]}")
```

```python
# To get one column on DataFrame

df[['Age']]
```

```python
# To get multiple columns on DataFrame

df[['Name','Salary']]
```

```python
# To drop a column from DataFrame

df.drop('Department', axis = 1) # axis = 1 indicates that we are dropping a column
```

```python
# To drop a row from DataFrame

df.drop(2, axis = 0) # axis = 0 indicates that we are dropping a row
```

```python
# # To drop both a column and a row from DataFrame

df.drop('Age',axis = 1).drop(3,axis = 0)
# Here first we are dropping the column 'Age' and then dropping the row with index 3
```

```python
# Original DataFrame prints

df
```

```python
df.drop(4,axis = 0, inplace = True)
# Here we are dropping the row with index 4 permanently from the original DataFrame
df
```
