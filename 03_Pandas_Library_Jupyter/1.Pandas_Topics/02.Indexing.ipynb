```python
# Indexing performed in Series of Pandas
```

```python
# single element access

# iloc only works with integer based indexing

import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])

print(f"Original Data:\n{data}")
print(f"1st Element: {data[0]}")
print(f"Element at index 3 with iloc : {data.iloc[3]}")
```

```python
# to access multiple elements, we can use slicing

import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])

print(f"Original Data: \n{data}")

# accessing elements from index 1 to 3 without iloc method

print(f"Elements from index 1 to 3:\n{data[1:4]}")

# Here the end index is not included

# if you want to access elements at 2 and 3 then

print(f"Elements at index 2 and 3 : \n{data[2:4]}")
```

```python
# getting elements using iloc operators

import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])
print(f"Original Series:\n{data}\n")

# for accessing single element using iloc

print(f"Element at index 2 using iloc: {data.iloc[2]}")

# for accessing multiple elements using iloc without range

print(f"Elements using iloc without range: \n{data.iloc[[1,2,4]]}\n")

# accessing elements using iloc with range

print(f"Elements from index 1 to 3 using iloc with range: \n{data.iloc[1:4]}")

# Changing value in a Series

data[2] = 60
print(f"Updated Series after changing value at index 4:\n{data}")
```

```python
# Changing the name of the Series and pass the index of new Series of fruits

data.name = "Calories"
index = pd.Series(["apple", "banana", "mango", "orange", "grape"])
data.index = index

print(f"Series with new index and name:\n{data}\n")

# Accessing elements

data["mango"]
print(f"Calories with Fruit 'mango': {data['mango']}")

print(f"Calories with Fruits 'banana' and 'grape': \n{data[['banana', 'grape']]}\n")
```

```python
# Getting elements using loc operator 
# In loc we can access elements using the index labels and last element is included in the output

import pandas as pd 
data = pd.Series([10, 20, 30, 40, 50])

data.name = "Calories"
index = pd.Series(["apple", "banana", "mango", "orange", "grape"])
data.index = index

print(f"Original Series:\n{data}\n")

# accessing single element using loc

print(f"Element with index label 'orange' using loc: {data.loc['orange']}\n")

# accessing multiple elements using loc

print(f"Elements from index label 'banana' to 'grape' using loc: \n{data.loc['banana':'grape']}\n")

```

```python
# Creating a series using dictionary

import pandas as pd

fruit_proteins = {
    "apple": 0.3,           # protein in grams
    "banana": 1.1,  
    "mango": 0.8,
    "orange": 0.9,
    "grape": 0.7,
    "kiwi": 1.0,
    "pineapple": 0.5,
    "peach": 0.6,
    "cherry": 1.0,
    "watermelon": 0.4
}

data = pd.Series(fruit_proteins,name="Proteins (g)")
print(f"Fruit Proteins Series:\n{data}\n")

# Accessing index and name of the Series with the help of loc

print(f"Gettig the Fruits from index 'mango' to 'peach' using loc: \n{data.loc['mango':'peach']}\n")
```
