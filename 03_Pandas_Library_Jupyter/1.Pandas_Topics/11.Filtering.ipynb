```python
import pandas as pd

df = pd.read_csv("C:\\Users\\dhamm\\Downloads\\archive\\athlete_events.csv")
df.head()
```

```python
df = df.loc[(df['Height']>215) & (df['NOC'].isin(["USA"])),["Name","Sport","Height","NOC"]]
df
```

```python
# Filtering DataFrame by the help of query method

df = df.query('Height > 215 and NOC == "USA" and Sport == "Basketball"')[["Name","Sport","Height","NOC"]]
df
```
