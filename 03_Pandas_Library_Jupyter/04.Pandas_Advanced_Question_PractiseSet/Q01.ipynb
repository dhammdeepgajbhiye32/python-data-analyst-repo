```python
import pandas as pd

df = pd.read_csv("C:\\Users\\dhamm\\Downloads\\archive\\athlete_events.csv")
df
```

```python
# Filtering the rows using multiple conditions with (& , | ) operators
df = df.loc[(df['Height'] > 214) & ((df['NOC'] =="CHN") | (df['NOC'] =="USA")) & ((df['Sport'] =="Basketball") | (df['Sport'] =="Volleyball")), ["Name","Sport","Height","NOC"]]
df
```
