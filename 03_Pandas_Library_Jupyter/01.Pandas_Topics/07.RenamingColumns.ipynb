```python
# To rename columns, we can use the 'rename' method in the dataframe,
# But to get changes reflected in the original dataframe, we need to set the 'inplace' parameter to True.


import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [70000, 80000, 60000, 90000],
    'Department': ['HR', 'Finance', 'IT', 'IT']
}

# Create a DataFrame from the dictionary

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
df
```

```python
# To rename columns, we can use the `rename` method of a pandas DataFrame.

df.rename(columns = {'Department': 'Dept'}, inplace = True)
df
```

```python
# To find the unique values in a column, we can use the `unique` method.

df['Dept'].unique()
```

```python
# To count the values in a specified column, we can use the `value_counts` method.

df['Dept'].value_counts()
```

```python
# To Create a new column known as 'Promoted Salary' by increasing the 'Salary' column by * 10.

df['Promoted Salary'] = df['Salary'] * 10
df
```
