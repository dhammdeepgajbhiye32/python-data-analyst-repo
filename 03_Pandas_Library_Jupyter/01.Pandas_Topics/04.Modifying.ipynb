```python
# Creating a series using dictionary

import pandas as pd

fruit_proteins = {
    "apple": 0.3,           # protein in grams
    "banana": 1.1,  
    "mango": 0.8,
    "orange": 0.9,
    "grape": 0.7,
    "kiwi": 1.0,
    "pineapple": 0.5,
    "peach": 0.6,
    "cherry": 1.0,
    "watermelon": 0.4
}

data = pd.Series(fruit_proteins, name = "Proteins")

# modifying the Series values

data["mango"] = 1.45
print(f"updated Series after modifying value at index 'mango':\n{data}\n")
```
